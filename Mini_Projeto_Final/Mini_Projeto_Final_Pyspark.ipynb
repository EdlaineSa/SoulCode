```python
# Abrindo o arquivo
file_location = "/content/drive/MyDrive/Mini Projeto Final/df_marketing_campaing_pandas.csv"
file_type = "csv"

# CSV 
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

customSchema = StructType([
    StructField("ID", IntegerType(), True),
    StructField("Ano_Nascimento", IntegerType(), True),
    StructField("Escolaridade", StringType(), True),
    StructField("Estado_Civil", StringType(), True),
    StructField("Renda", DoubleType(), True),
    StructField("Crianca", IntegerType(), True),
    StructField("Adolescente", IntegerType(), True),
    StructField("Data_Cadastro", DateType(), True),
    StructField("Dias_Ultima_Compra", IntegerType(), True),
    StructField("Qtd_Vinho", IntegerType(), True),
    StructField("Qtd_Frutas", IntegerType(), True),
    StructField("Qtd_Prod_Carne", IntegerType(), True),
    StructField("Qtd_Prod_Peixe", IntegerType(), True),
    StructField("Qtd_Prod_Doce", IntegerType(), True),
    StructField("Qtd_Prod_Gold", IntegerType(), True),
    StructField("Num_Compras_Promocao", IntegerType(), True),
    StructField("Num_Compras_Site", IntegerType(), True),
    StructField("Num_Compras_Catalogo", IntegerType(), True),
    StructField("Num_Compras_Loja", IntegerType(), True),
    StructField("Num_Compras_Site_Mensal", IntegerType(), True),
    StructField("Campanha_Aceita_3", StringType(), True),
    StructField("Campanha_Aceita_4", StringType(), True),
    StructField("Campanha_Aceita_5", StringType(), True),
    StructField("Campanha_Aceita_1", StringType(), True),
    StructField("Campanha_Aceita_2", StringType(), True),
    StructField("Reclamacoes", StringType(), True),
    StructField("Resposta", StringType(), True),
])

df = spark.read.load(file_location, format="csv", header="true", sep=',', schema=customSchema)

df.show(5)
df.printSchema()
```

```python
#Realizar a mudança de nome de pelo menos 2 colunas
df = df.withColumnRenamed("ID", "Id_Cliente").withColumnRenamed("Resposta", "Resposta_Cliente")
df.printSchema()
```

```python
#Deverá criar pelo menos duas novas colunas contendo alguma informação relevante 
#sobre as outras colunas já existentes(Use a sua capacidade analítica)
df = df.withColumn("Filhos_Total", F.col("Crianca") + F.col("Adolescente"))
df = df.withColumn("Qtd_Comprada_Total", F.col("Qtd_Vinho") + F.col("Qtd_Frutas") + F.col("Qtd_Prod_Carne") + F.col("Qtd_Prod_Peixe") + F.col("Qtd_Prod_Doce") + F.col("Qtd_Prod_Gold"))
df = df.withColumn("Idade_Cliente", 2021 - F.col("Ano_Nascimento"))

df.head(5)
```

```python
# Deverá utilizar filtros, ordenação e agrupamento, 
# trazendo dados relevantes para o negócio em questão.

#Soma de renda de acordo com Escolaridade
df.groupby(F.col("Escolaridade")).sum("Renda").alias("Soma_Renda").show()
```

```python
#Verificando quantos doutores possuem 1 ou mais filhos

df.filter((F.col("Escolaridade") == "Doutorado") & (F.col("Filhos_Total") >= 1)).show()
```

```python
#Somatório, Media, Maior e Menor Renda de acordo com a Escolaridade
(df.groupby(F.col("Escolaridade")).agg(
            F.round(F.sum("Renda"), 2).alias("Soma_Renda"),
            F.round(F.mean("Renda"), 2).alias("Media_Renda"),
            F.max("Renda").alias("Maior_Renda"),
            F.min("Renda").alias("Menor_Renda")
            ).orderBy("Maior_Renda").show())
```

```python
#Soma e media de Idade de acordo com a escolaridade
df.groupby(F.col("Escolaridade")).agg(
            F.sum("Idade_Cliente").alias("Soma_Idade"),
            F.round(F.mean("Idade_Cliente"),2).alias("Media_Idade"),
            ).show()
```

```python
#Utilizar pelo menos duas Window Functions
```

```python
w0 = Window.orderBy(F.col("Idade_Cliente"))
df = df.withColumn("Ordem", F.row_number().over(w0))
df.select(F.col("Ordem"), F.col("Idade_Cliente"), F.col("Id_Cliente"), F.col("Estado_Civil"), F.col("Filhos_Total"), F.col("Qtd_Comprada_Total")).show(25)
```

```python
df.withColumn("Percent_Rank", F.percent_rank().over(w0)).show()
```

```python
w1 = Window.partitionBy(F.col("Estado_Civil")).orderBy(F.col("Filhos_Total"))
df = df.withColumn("Rank", F.row_number().over(w1))
df.select(F.col("Rank"), F.col("Renda"), F.col("Num_Compras_Promocao"), F.col("Num_Compras_Catalogo"), F.col("Num_Compras_Loja"), F.col("Num_Compras_Site_Mensal"), F.col("Qtd_Comprada_Total")).show(15)
```
