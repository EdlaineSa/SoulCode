```python
import pandas as pd
```

```python
#Carregando o arquivo e exibindo (as primeiras 20 linhas)
df = pd.read_csv("/content/drive/MyDrive/Mini Projeto Final/marketing_campaign.csv - marketing_campaign.csv.csv", sep=",")
df.head(5)
```

```python
#Verificando os valores NaN
#Apagando os valores ausentes encontrados (já que são poucas linhas)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Verificando dados da coluna "Z_CostContact"
print(df['Z_CostContact'].unique())
```

```python
#Verificando dados da coluna "Z_Revenue"
print(df['Z_Revenue'].unique())
```

```python
#Já que os valores das colunas "Z_CostContact" e "Z_Revenue"se repetem, 
#não apresentam dados diferentes ao Dataframe, assim é possivel deletar as
#colunas sem perda de dados para comparação
colunas = ['Z_CostContact', 'Z_Revenue']
df.drop(colunas, axis=1, inplace=True)
df.head(5)
```

```python
#Transformando "Dt_Customer" de "object" para "Data"
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df.dtypes
```

```python
#Verificando os dados nas colunas tipo "object"
print(df['Education'].unique())
```

```python
#Verificando os dados nas colunas tipo "object"
print(df['Marital_Status'].unique())
```

```python
#TRADUÇÃO
#Renomeando as colunas
df.rename(columns= {"Year_Birth":"Ano_Nascimento", 
    "Education":"Escolaridade", 
    "Marital_Status":"Estado_Civil", 
    "Income":"Renda", 
    "Kidhome":"Crianca", 
    "Teenhome":"Adolescente", 
    "Dt_Customer":"Data_Cadastro", 
    "MntWines":"Qtd_Vinho", 
    "MntFruits":"Qtd_Frutas", 
    "MntMeatProducts":"Qtd_Prod_Carne", 
    "MntFishProducts":"Qtd_Prod_Peixe", 
    "MntSweetProducts":"Qtd_Prod_Doce",
    "MntGoldProds":"Qtd_Prod_Gold", 
    "NumDealsPurchases":"Num_Compras_Promocao", 
    "NumWebPurchases":"Num_Compras_Site", 
    "NumCatalogPurchases":"Num_Compras_Catalogo", 
    "NumStorePurchases":"Num_Compra_Loja", 
    "NumWebVisitsMonth":"Num_Visitas_Site_Mensal", 
    "AcceptedCmp3":"Campanha_Aceita_3", 
    "AcceptedCmp4":"Campanha_Aceita_4", 
    "AcceptedCmp5":"Campanha_Aceita_5", 
    "AcceptedCmp1":"Campanha_Aceita_1", 
    "AcceptedCmp2":"Campanha_Aceita_2", 
    "Complain":"Reclamacoes", 
    "Recency":"Dias_Ultima_Compra",
    "Response":"Resposta"},
    inplace=True
    )
df.head(5)
```

```python
#TRADUÇÃO
#Renomeando as linhas da coluna "Estado Civil"
df['Estado_Civil'].replace('Single', 'Solteiro(a)', inplace=True)
df['Estado_Civil'].replace('Together', 'Namorando', inplace=True)
df['Estado_Civil'].replace('Married', 'Casado(a)', inplace=True)
df['Estado_Civil'].replace('Divorced', 'Divorciado(a)', inplace=True)
df['Estado_Civil'].replace('Widow', 'Viuvo(a)', inplace=True)
df['Estado_Civil'].replace(['Alone', 'Absurd', 'YOLO'], pd.NA, inplace=True)
print(df['Estado_Civil'].unique())
```

```python
#TRADUÇÃO
#Renomeando as linhas da coluna "Escolaridade"
df['Escolaridade'].replace('Graduation', 'Graducacao', inplace=True)
df['Escolaridade'].replace('PhD', 'Doutorado', inplace=True)
df['Escolaridade'].replace('Master', 'Mestrado', inplace=True)
df['Escolaridade'].replace('Basic', 'Ensino Fundamental', inplace=True)
df['Escolaridade'].replace('2n Cycle', 'Ensino Medio', inplace=True)
print(df['Escolaridade'].unique())
```

```python
#Verificando dados das colunas "Campanha_Aceita_1", "Campanha_Aceita_2", "Campanha_Aceita_3", 
#"Campanha_Aceita_4", "Campanha_Aceita_5", "Reclamacoes" e "Respostas", é possivel substituir as rrespostas 
#cpara "sim" ou não" para melhor compreensão

print(df['Campanha_Aceita_1'].unique())
print(df['Campanha_Aceita_2'].unique())
print(df['Campanha_Aceita_3'].unique())
print(df['Campanha_Aceita_4'].unique())
print(df['Campanha_Aceita_5'].unique())
print(df['Reclamacoes'].unique())
print(df['Resposta'].unique())
```

```python
#Substituindo
df['Resposta'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Reclamacoes'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Campanha_Aceita_1'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Campanha_Aceita_2'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Campanha_Aceita_3'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Campanha_Aceita_4'].replace({0:'Não', 1:"Sim"}, inplace=True)
df['Campanha_Aceita_5'].replace({0:'Não', 1:"Sim"}, inplace=True)

print(df['Campanha_Aceita_1'].unique())
print(df['Campanha_Aceita_2'].unique())
print(df['Campanha_Aceita_3'].unique())
print(df['Campanha_Aceita_4'].unique())
print(df['Campanha_Aceita_5'].unique())
print(df['Resposta'].unique())
```
